```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt


x = np.loadtxt('colo1.txt')

num_bins = 50

fig, ax = plt.subplots()

# the histogram of the data
n, bins, patches = ax.hist(x, num_bins, density=True)



#ax.plot(bins, y, '--')
ax.set_xlabel(r'KE salt bridge length in $\AA$',fontsize=18)
ax.set_ylabel('Frequency density',fontsize=18)
ax.set_title('KE salt bridge histogram',fontsize=18)
plt.xlim((0,25))
plt.xticks(fontsize= 10)
plt.yticks(fontsize= 20)
# Tweak spacing to prevent clipping of ylabel
fig.tight_layout()
plt.savefig('histzone1.png', bbox_inches = 'tight', dpi = 300)
plt.show()
```
